```python
from sklearn import svm
from sklearn.svm import SVR
import numpy as np
import pandas as pd
 

data = pd.read_csv('test.csv',delimiter=',', header = None)
y=data[0].to_numpy()
del data[0]
X=data.to_numpy()
regressor = SVR(kernel='linear', epsilon=0.1,C=1.5 )
regressor.fit(X,y)
print(regressor.coef_)
print(regressor.intercept_)
```
